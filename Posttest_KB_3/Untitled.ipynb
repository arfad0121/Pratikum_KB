```python
df = df.dropna(subset=['Type'])
```

```python
pie = df['Type'].value_counts().rename_axis('Type').reset_index(name='Rating')
```

```python
plt.figure(figsize=(10,10))
plt.pie(pie.Rating, labels=pie.Type, startangle=90,autopct='%.1f%%')
plt.title("Rating")
plt.show()
```

```python
melihat grafik pie diatas banyak sekali game yang free dibandingkan dengan paid dan 0
```
