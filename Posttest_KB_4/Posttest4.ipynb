Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

```python
imputer = SimpleImputer(strategy = 'mean')
df["Rating"] = imputer.fit_transform(df[["Rating"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["Type"] = imputer.fit_transform(df[["Type"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

Mengubah Data Menjadi Katagorikal

```python
from sklearn.preprocessing import OrdinalEncoder 
```

```python
encoder = OrdinalEncoder() 
df["Genres"] = encoder.fit_transform(df[["Category"]]) 
df.head()
```

One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
encoder = OneHotEncoder(sparse=False) 
maritalEnc = encoder.fit_transform(df[["Content Rating"]]) 
maritalEnc = pd.DataFrame(maritalEnc) 
df = df.join(maritalEnc) 
df 
```
